```python
import numpy as np
```

```python
x = np.linspace(0,2,10)
y = np.linspace(0,2,10)
#y=X**2
x
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(x,y, label='quadratique')  # C= couleur du trait, ls= style du trait, label=nom de la courbe, lw=epaisseur du trait
plt.plot(x, x**3, label='cubique') 
plt.title('Figure 1') 
plt.xlabel('axe x') 
plt.ylabel('axe y') 
plt.legend()
plt.show()
```

```python
plt.scatter(x,y) # Nuage de points
plt.show()
```
