               1.  Pyplot

```python
import numpy as np
```

```python
x = np.linspace(0,2,10)
y = np.linspace(0,2,10)
#y=X**2
x
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.plot(x,y, label='quadratique')  # C= couleur du trait, ls= style du trait, label=nom de la courbe, lw=epaisseur du trait
plt.plot(x, x**3, label='cubique') 
plt.title('Figure 1') 
plt.xlabel('axe x') 
plt.ylabel('axe y') 
plt.legend()
plt.savefig("figure.png")
plt.show()
```

```python
plt.scatter(x,y) # Nuage de points
plt.show()
```

                   SUBPLOT

```python
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,y, c='red')
plt.subplot(2,1,2)
plt.plot(x,y, c='blue')
```

                                 TYPES DE GRAPHIQUES

```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import load_iris
```

```python
iris= load_iris()
x= iris.data
y=iris.target
names= list(iris.target_names)

print(f"x contient {x.shape[0]} exemples et {x.shape[1]} variables")
print(f"il y a {np.unique(y).size} classes")
y
```

                       1. GRAPHIQUE AVEC PLT.SCATTER

```python
plt.scatter(x[:,0], x[:,1],c=y, alpha=0.5, s=100) # alpha= c'est pour controler la transparence, s= taille de points
plt.xlabel("longueur de sepal")
plt.ylabel("largeur de sepal")
```

                      2. GRAPHIQUE 3D

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
ax = plt.axes(projection='3d')
ax.scatter(x[:,0], x[:,1],x[:,2], c=y)
```

                        3. HISTOGRAMME  # visualiser les distribution

```python
plt.hist(x[:,0], bins=30) 
plt.hist(x[:,1], bins=30) #bins=determine les nombre de section
```

```python
plt.hist2d(x[:,0],x[:,1], cmap="Blues") 
plt.xlabel("longueur de sepal")
plt.ylabel("largeur de sepal")
plt.colorbar()
```

```python
from scipy import misc
face=misc.face(gray= True)
# plt.imshow(face,cmap='gray')
plt.hist(face.ravel(),bins=255)
plt.show()
```
